## modify report data based on raw data

```python
def modify_report_data(report_df, raw_df):
    # Create a dictionary from raw_df for faster access
    raw_dict = defaultdict(list)
    for i, row in raw_df.iterrows():
        raw_dict[row['LASID']].append(row)

    # Prepare dictionaries to store the changes we want to make
    changes_school_system_code = {}
    changes_school_system_name = {}
    changes_lasid = {}

    # Iterate over the rows in report_df
    for index, report_row in report_df.iterrows():
        lasid = report_row['LASID']
        # Check if the LASID exists in raw_df
        if lasid in raw_dict:
            # Get the corresponding rows in the raw_df DataFrame
            raw_rows = raw_dict[lasid]
            # Iterate over the raw_rows
            for raw_row in raw_rows:
                if str(raw_row['Site_Report_Code']).strip() == 'R36' and (raw_row[f'Test_Taken_Flag_-_{subject}'] == 'Y' or raw_row[f'Summarized_Void_Flag_-_{subject}'] == 'Y' or raw_row[f'Launch_Flag_-_{subject}'] == 'Y'):
                    changes_school_system_code[index] = 'R36'
                    changes_school_system_name[index] = 'Orleans Parish'
                if (str(raw_row['School_Code']).strip() in ['998', '997'] or str(raw_row['Assess_Group_Type']).strip() in ['13', '05']) and (raw_row[f'Test_Taken_Flag_-_{subject}'] == 'Y' or raw_row[f'Summarized_Void_Flag_-_{subject}'] == 'Y' or raw_row[f'Launch_Flag_-_{subject}'] == 'Y'):
                    changes_lasid[index] = 'XXXXX' + str(report_row['LASID'])[5:]

    # Apply the changes to the DataFrame
    for index, value in changes_school_system_code.items():
        report_df.at[index, 'School_System_Code'] = value
    for index, value in changes_school_system_name.items():
        report_df.at[index, 'School_System_Name'] = value
    for index, value in changes_lasid.items():
        report_df.at[index, 'LASID'] = value

    return report_df

# Use the function to modify the report_df DataFrame
report_df = modify_report_data(report_df, raw_df)
```

# 1. Calculate the average scores based on pdf report informations

```python
# Define a function to round up if the number is halfway between two others
def round_half_up(x):
    x = Decimal(x).quantize(Decimal('0.00'))  # Control the precision of the division operation
    decimal_part = x - Decimal(int(x))
    if abs(decimal_part - Decimal('0.5')) < Decimal('1e-10'):
        return Decimal(ceil(x))
    else:
        return Decimal(round(x))



report_systems = report_df[['School_System_Code', 'School_System_Name']].drop_duplicates()

total_systems = raw_df['School_System_Name'].unique()

total_system_codes = raw_df['School_System_Code'].unique()

r_system_df = pd.DataFrame(report_systems)


r_system_df['Name_Is_Valid'] = r_system_df['School_System_Name'].apply(lambda x: x.strip() in ''.join(total_systems))

r_system_df['Code_Is_Valid'] = r_system_df['School_System_Code'].astype(str).apply(lambda x: x.strip() in ' '.join(total_system_codes))

# Convert the scores to numeric values
raw_df.loc[:, f"Scale_Score_-_{subject}"] = pd.to_numeric(raw_df[f"Scale_Score_-_{subject}"], errors='coerce')

# Filter the DataFrame for rows where 'Summarized_Void_Flag_-_{subject}' is not equal to 2
filtered_raw_df = raw_df[~raw_df[f'Summarized_Void_Flag_-_{subject}'].str.strip().isin(['2', '1', '4'])]



# Calculate the school system average for each grade
system_grade_avg = filtered_raw_df.groupby(['School_System_Name', 'Summarized_Grade'])[f"Scale_Score_-_{subject}"].apply(lambda x: round_half_up((Decimal(x.sum()) / Decimal(len(x))).quantize(Decimal('0.00'))))

# Convert the series to a dataframe
system_grade_avg_df = system_grade_avg.reset_index()

# Rename the columns
system_grade_avg_df.columns = ['School_System_Name', 'Grade', f'Average_Scale_Score_{subject}']

# Filter the DataFrame for rows where 'Rollup_to_State' is 'Y'
state_df = raw_df[raw_df['Rollup_to_State'] == 'Y']

# Calculate the state average for each grade
state_grade_avg = state_df.groupby('Summarized_Grade')[f"Scale_Score_-_{subject}"].apply(lambda x: round_half_up((Decimal(int(x.dropna().sum())) / Decimal(int(x.dropna().count()))).quantize(Decimal('0.00'))))

# Convert the series to a dataframe and append it to the system_grade_avg_df dataframe
state_grade_avg_df = state_grade_avg.reset_index()
state_grade_avg_df['School_System_Name'] = 'State'
state_grade_avg_df.columns = ['Grade', f'Average_Scale_Score_{subject}', 'School_System_Name']

# Append state_grade_avg_df to system_grade_avg_df
system_grade_avg_df = pd.concat([system_grade_avg_df, state_grade_avg_df], ignore_index=True)

```

## calculate school levels

```python
# Define the achievement levels to include in the calculation
achievement_levels = ['ADV', 'MAS', 'BAS', 'APP', 'UNS']

# Filter the raw_df DataFrame
filtered_raw_df = raw_df[raw_df['School_Name'].str.strip().isin(r_school_df['School_Name']) & raw_df[f'Assessment_Achievement_Level_-_{subject}'].str.strip().isin(achievement_levels)]

# Calculate the counts for each school, grade, and subject
counts_df = filtered_raw_df.groupby(['School_Name', 'Summarized_Grade', f'Assessment_Achievement_Level_-_{subject}']).size().unstack(fill_value=0)

# Calculate the percentage of students at each achievement level for each school, grade, and subject
school_grade_subject_level_df = counts_df.div(counts_df.sum(axis=1), axis=0).applymap(lambda x: Decimal(str(x)))

# Calculate the total count for each (School_Name, Summarized_Grade) group
total_counts_df = counts_df.sum(axis=1)


# Define the format_percentage function to take an additional total_count parameter
def format_percentage(x, total_count):
    if total_count < 10:
        return 'NR'
    elif x < Decimal('0.01') or round(x * 100) == 1:
        return '≤1%'
    else:
        percentage = x * 100
        return f'{percentage.quantize(Decimal("1"), rounding=ROUND_HALF_UP)}%'

# Apply the function to the dataframe
formatted_school_grade_subject_level_df = school_grade_subject_level_df.apply(lambda x: x.apply(format_percentage, total_count=total_counts_df.loc[x.name]), axis=1)
```

```python
def convert_to_str(value):
    try:
        # Try to convert to int first to remove decimal points
        return str(int(value))
    except ValueError:
        # If it's not a number, just convert to string
        return str(value)
```

```python
# Keep only the rows in raw_df where the School_Name is in r_school_names
raw_audit_df = raw_audit_df[raw_audit_df['School_Name'].isin(r_school_names)]

start_time = time.time()
# Define a dictionary to map the full names to the abbreviations
level_mapping = {
    'Advanced': 'ADV',
    'Mastery': 'MAS',
    'Basic': 'BAS',
    'Approaching_Basic': 'APP',
    'Unsatisfactory': 'UNS'
}

# Define a dictionary to map the raw data values to the report data values
rating_mapping = {
    '1': 'WEAK PERFORMANCE',
    '2': 'MODERATE PERFORMANCE',
    '3': 'STRONG PERFORMANCE',
    '': ''
    
}



# Create an empty list to store the discrepancies
discrepancies = []

# Convert LASID in raw_df to a set for faster lookup
raw_lasid_set = set(raw_df['LASID'])



# Create a dictionary from raw_df for faster access
raw_dict = defaultdict(list)
for i, row in raw_audit_df.iterrows():
    raw_dict[row['LASID']].append(row)


# Record the end time
end_time = time.time()

# Calculate and print the running time
running_time = end_time - start_time
print(f"Running time: {running_time} seconds")
```

```python
discrepancies = []

start_time = time.time()
# Iterate over the rows in report_df
for index, report_row in tqdm(report_df.iterrows(), total=report_df.shape[0]):
    lasid = report_row['LASID']
    # Check if the LASID exists in raw_df
    if lasid not in raw_lasid_set:
        continue

    # Get the corresponding rows in the raw_df DataFrame
    raw_rows = raw_dict[lasid]
    
    # Iterate over the raw_rows
    for raw_row in raw_rows:
        
        # Check if the ELA score matches
        raw_value = pd.to_numeric(raw_row[f'Scale_Score_-_{subject}'], errors='coerce')
        report_value = pd.to_numeric(report_row[f'Scale_Score_-_{subject}'], errors='coerce')
        if raw_value != report_value:
            continue

   
        # Compare the personal information
        for column in ['Student_First_Name', 'Student_Last_Name', 'Summarized_DOB_Month', 
                       'Summarized_DOB_Day', 'Summarized_DOB_Year', 'Summarized_Grade', 'School_Name', 
                       'School_Code', 'School_System_Name', 'School_System_Code']:
           # Then in your comparison code, use this function to convert values to strings:
            report_value = convert_to_str(report_row[column]).strip()
            raw_value = convert_to_str(raw_row[column]).strip()
            if pd.isnull(report_value) and pd.isnull(raw_value):
                continue
            elif column in ['Summarized_DOB_Month', 'Summarized_DOB_Day', 'Summarized_DOB_Year']:
                # If the raw data is empty, skip the comparison
                if raw_value == '':
                    continue
                # Convert both values to integers before comparing
                if pd.to_numeric(report_value, errors='coerce') != pd.to_numeric(raw_value, errors='coerce'):
                    discrepancy = f"{column} does not match"
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
            elif column == 'Student_First_Name':
                # Check if the raw data matches the first letter of the first name
                if raw_value == report_value[0]:
                    continue
                if raw_value != report_value[0]:
                    if raw_value != report_value:
                        if raw_value != report_value[:8]:
                            discrepancy = f"{column} does not match"
                            discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                                  'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
            elif column == 'Student_Last_Name':
                # Check if the raw data matches the first three letters of the last name
                if raw_value == report_value[0]:
                    continue
                if raw_value != report_value[:3]:
                    if raw_value != report_value:
                        if raw_value != report_value[:12]:
                            discrepancy = f"{column} does not match"
                            discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                                  'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
            elif column == 'School_System_Code':
                # Check if the raw data is a zero-padded version of the report data
                if raw_value != report_value and raw_value.lstrip('0') != report_value:
                    discrepancy = f"{column} does not match"
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
            else:
                if report_value != raw_value:
                    discrepancy = f"{column} does not match"
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
        # Compare the scores

        # Check if the raw data for this subject is valid
        raw_value = pd.to_numeric(raw_row[f'Scale_Score_-_{subject}'], errors='coerce')
        report_value = pd.to_numeric(report_row[f'Scale_Score_-_{subject}'], errors='coerce')
        if pd.isnull(report_value) and pd.isnull(raw_value):
            continue
        elif report_value != raw_value:
            discrepancy = f"{subject} score does not match"
            discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                  'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})

        # Compare the additional columns
        for column_raw, column_report in [('Reading_Rating_', 'Reading_Performance_Achievement_Level'), 
                                          ('Literary_Text_Rating', 'Literary_Text_Achievement_Level'), 
                                          ('Informational_Text_Rating', 'Informational_Text_Achievement_Level'), 
                                        ('Vocabulary_Rating', 'Vocabulary_Achievement_Level'), 
                                          ('Written_Expression_Rating', 'Written_Expression_Achievement_Level'), 
                                          ('Knowledge_&_Use_of_Language_Conventions_Rating', 'Knowledge&Use_of_Language_Conventions'), 
                                          ('Writing_Rating', 'Writing_Performance_Achievement_Level'), 
                                         ]:
            raw_value = raw_row[column_raw]
            report_value = report_row[column_report]
            # Convert the raw value using the mapping
            raw_value_converted = rating_mapping.get(raw_value, 'Unknown')
            # If raw_value_converted is blank and report_value is 'NaN' or blank, continue to the next iteration
            if (raw_value.strip() == '' or pd.isna(raw_value) ) and (str(report_value).lower().strip() == 'nan' or report_value.strip() == '' or pd.isna(report_value)):
                continue
            elif raw_value_converted.strip() != report_value.strip():
                discrepancy = f"{column_report} does not match"
                discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                      'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})
#                print(raw_row[f'Test_Taken_Flag_-_{subject}'], raw_row[f'Summarized_Void_Flag_-_{subject}'], raw_row[f'Launch_Flag_-_{subject}'])


        # Compare the average scores
        system_grade_avg_row = system_grade_avg_df[(system_grade_avg_df['School_System_Name'] == report_row['School_System_Name']) & 
                                                   (system_grade_avg_df['Grade'] == report_row['Summarized_Grade'])]
        if not system_grade_avg_row.empty:
            system_grade_avg_row = system_grade_avg_row.iloc[0]
            report_value = report_row['School_System_Average_Score']
            # Check if the report_value is NaN before proceeding
            if pd.isnull(report_value):
                continue
            raw_value = system_grade_avg_row[f'Average_Scale_Score_{subject}']
            if pd.isnull(report_value) and pd.isnull(raw_value):
                continue
            elif report_value != raw_value:
                discrepancy = f"{subject} school system average score does not match"
                discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                      'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})

        # Compare the state average scores
        state_avg_row = system_grade_avg_df[(system_grade_avg_df['School_System_Name'] == 'State') & 
                                            (system_grade_avg_df['Grade'] == report_row['Summarized_Grade'])]
        if not state_avg_row.empty:
            state_avg_row = state_avg_row.iloc[0]
            report_value = report_row['State_Average_Score']
            raw_value = state_avg_row[f'Average_Scale_Score_{subject}']
            if pd.isnull(report_value) and pd.isnull(raw_value):
                continue
            elif report_value != raw_value:
                discrepancy = f"{subject} State average score does not match"
                discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                      'Discrepancy': discrepancy, 'Report_Data': report_value, 'Raw_Data': raw_value})

        # Use the mapping in the loop
        for level in ['Advanced', 'Mastery', 'Basic', 'Approaching_Basic', 'Unsatisfactory']:
            # Get the corresponding row from the level DataFrame
            level_row = formatted_school_grade_subject_level_df.loc[[(raw_row['School_Name'], raw_row['Summarized_Grade'])]]
            level_row_system = formatted_school_system_grade_subject_level_df.loc[[(raw_row['School_System_Name'], raw_row['Summarized_Grade'])]]
            level_row_state = formatted_state_grade_subject_level_df.loc[[raw_row['Summarized_Grade']]]
            
            # Get the abbreviation for the level
            level_abbrev = level_mapping[level]



            # Calculate the difference between the report data and the raw data for this level
            if pd.isna(report_row[f'{level}_School_Percentage']) and pd.isna(level_row[level_abbrev].iloc[0]):
                continue
            else:
                # Compare the percentage strings directly
                report_percentage = report_row[f'{level}_School_Percentage']
                level_percentage = level_row[level_abbrev].iloc[0]
                if report_percentage != level_percentage:
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': f"{level} school percentage does not match", 
                                          'Report_Data': report_row[f'{level}_School_Percentage'], 'Raw_Data': level_row[level_abbrev].iloc[0]})

            # Calculate the difference between the report data and the raw data for this level
            if pd.isna(report_row[f'{level}_School_System_Percentage']) and pd.isna(level_row_system[level_abbrev].iloc[0]):
                continue
            else:
                # Compare the percentage strings directly
                report_percentage_system = report_row[f'{level}_School_System_Percentage']
                level_percentage_system = level_row_system[level_abbrev].iloc[0]
                if report_percentage_system != level_percentage_system:
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': f"{level} school system percentage does not match", 
                                          'Report_Data': report_row[f'{level}_School_System_Percentage'], 'Raw_Data': level_row_system[level_abbrev].iloc[0]})

            if pd.isna(report_row[f'{level}_State_Percentage']) and pd.isna(level_row_state[level_abbrev].iloc[0]):
                continue
            else:
                # Compare the percentage strings directly
                report_percentage_state = report_row[f'{level}_State_Percentage']
                level_percentage_state = level_row_state[level_abbrev].iloc[0]
                if report_percentage_state != level_percentage_state:
                    discrepancies.append({'LASID': lasid, 'Student_Name': report_row['Student_Name'], 
                                          'Discrepancy': f"{level} state percentage does not match", 
                                          'Report_Data': report_row[f'{level}_State_Percentage'], 'Raw_Data': level_row_state[level_abbrev].iloc[0]})


        
        
# Convert the list of discrepancies to a DataFrame
discrepancies_df = pd.DataFrame(discrepancies)

# Record the end time
end_time = time.time()

# Calculate and print the running time
running_time = end_time - start_time
print(f"Running time: {running_time} seconds")

# Print the discrepancies
discrepancies_df
```

```python
# Create a new list to store the updated discrepancies
updated_discrepancies = []

# Iterate over the discrepancies
for discrepancy in discrepancies:
    # Get the LASID
    lasid = discrepancy['LASID']
    
    # Get the corresponding row in the raw_df DataFrame
    raw_row = raw_df[raw_df['LASID'] == lasid].iloc[0]
    
    # Check if the 'Show_'N/A'_in_place_of_School_System_results_on_school_reports' field is 'Y'
    if raw_row["Show_'N/A'_in_place_of_School_System_results_on_school_reports"] == 'Y':
        # Check if the Report_Data value for this discrepancy is NaN
        if pd.isnull(discrepancy['Report_Data']):
            # If it is, skip this discrepancy
            continue
            
        
    
    # If the discrepancy wasn't skipped, add it to the updated discrepancies list
    updated_discrepancies.append(discrepancy)

# Convert the updated discrepancies list to a DataFrame
updated_discrepancies_df = pd.DataFrame(updated_discrepancies)

# Print the updated discrepancies
updated_discrepancies_df
```
